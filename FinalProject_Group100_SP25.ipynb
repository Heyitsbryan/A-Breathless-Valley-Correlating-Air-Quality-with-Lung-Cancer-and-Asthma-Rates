## Background and Prior Work

**Background**

The Central Valley of California is a region known not only for its agricultural productivity but also for its persistent air quality issues. Enclosed by mountain ranges and densely populated with farming and industrial activity, the area often experiences high concentrations of air pollutants, such as fine particulate matter (PM2.5). The matter accumulates in the small airways of the lungs, and irritants cause genetic changes responsible for both benign and malignant long-term lung diseases <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1).

Over the course of the decade, more research has been conducted on the potential connection between deteriorating air quality and increased respiratory health problems among residents, including allergy, asthma, and lung cancer diagnoses, especially in Sacramento Valley and San Joaquin Valley <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). From 2015 to 2023, concerns have grown about how this compromised air quality may be affecting public health, particularly the prevalence of seasonal allergies (formally known as allergic rhinitis), asthma, and lung cancer.

**Previous work**

"The Growing Impact of Air Quality on Lung-Related Illness" published in the Journal of Thoracic Disease in 2023, synthesizes findings from studies conducted between 1992 and 2023. The article’s key findings include the effects of particulate matter (PM2.5) on lung function, as epigenetic changes are primarily responsible for benign and malignant lung diseases. The article also found that urban areas experience worse air quality due to higher industrial activity. The article only discusses air quality as a general topic, but this project aims to look at the connection between elevation and air quality.

Meng, Wilhelm, Lombardi, Balmes, and Ritz (2010)’s "Outdoor Air Pollution and Uncontrolled Asthma in the San Joaquin Valley, California" examines the link between air pollution and asthma in the region. According to the California Health Interview Survey, around 1,502 San Joaquin Valley individuals reported having asthma. They discovered that those living in high-pollution areas were more likely to have asthma symptoms, and that poor air quality exacerbated asthma symptoms and increased the number of asthma diagnoses. Overall, it supports our focus on air quality and the effects it has on people's health, assisting us in explaining why many migrate away from these locations since it has only caused them damage, and their desire to get away from it. However, the findings of the study only relate to asthma. The project continues the study by generalizing findings to other respiratory ailments too, such as allergic rhintis and lung cancer.

In general, asthma prevalence in the Central Valley has increased from the period of 2015-2016 to 2021-2022. There are three outliers in the data; let's take a look at them.

```python
def find_outliers(s):
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
    return s[(s < lower) | (s > upper)]

outlier_list = []
for year, group in chis_data.groupby('start_year'):
    out_s = find_outliers(group['current_prevalence'])
    counties = group.loc[out_s.index, 'county']
    for cnt, val in zip(counties, out_s):
        outlier_list.append({'start_year': year, 'county': cnt, 'prevalence': val})

outliers_df = pd.DataFrame(outlier_list)
print(outliers_df)
```

Kings and Kern counties have some of the highest asthma prevalence rates in the Central Valley, with Kings County having a rate of 15.8% and Kern County having a rate of 16.7%.

### Asthma Prevalence in the Central Valley Over Time

Now, let's look at the asthma data and compare the prevalence of asthma across different counties in the Central Valley. First, we filter chis_data and pm_data for the counties in the Central Valley they share.

```python
chis_counties = chis_data["county"].unique()
pm_counties = pm_data["location"].unique()

# find the intersection of the two lists
common_counties = set(pm_counties).intersection(set(chis_counties))
print("Common counties between PM data and CHIS data:")
print(common_counties)
```

In pm_data, we are given the "timeframe" of when the PM2.5 levels were measured, and in chis_data, we are given a two-year interval of when the asthma data was collected. For the sake of simplicity, we will use the first year of the two-year interval in chis_data to match with the timeframe in pm_data.

```python
chis_years = chis_data["start_year"].unique().tolist()
pm_years = pm_data["timeframe"].unique().tolist()

# find the intersection of the two lists
common_years = set(pm_years).intersection(set(chis_years))
print("Common years between PM data and CHIS data:")
print(common_years)
```

This filters both datasets to keep only the matching counties and years. It prepares the data for clean merging and analysis.

```python
common_chis_data = chis_data[chis_data["county"].isin(common_counties) & chis_data["start_year"].isin(common_years)]
common_pm_data = pm_data[pm_data["location"].isin(common_counties) & pm_data["timeframe"].isin(common_years)]
```

This isolates the asthma and PM2.5 data, specifically for Sacramento County. A more focused case study to observe local patterns in air quality and asthma over time.

```python
# lets looks at sacramento
sacramento_chis = common_chis_data[common_chis_data["county"] == "sacramento"]
sacramento_pm = common_pm_data[common_pm_data["location"] == "sacramento"]

display(sacramento_chis)
display(sacramento_pm)
```

We'll then create a plot that shows asthma prevalence over time in Sacramento County, including error bars for confidence intervals. It helps assess how asthma rates have shifted across two-year intervals with uncertainty.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sacramento_chis['period'] = (sacramento_chis['start_year'].astype(str)+ '-' + sacramento_chis['end_year'].astype(str))

plt.figure(figsize=(8, 5))
sns.lineplot(
    data=sacramento_chis,
    x='period',
    y='current_prevalence',
    marker='o'
)

xpos = range(len(sacramento_chis))
plt.errorbar(
    xpos,
    sacramento_chis['current_prevalence'],
    yerr=[
        sacramento_chis['current_prevalence'] - sacramento_chis['ci_lower'],
        sacramento_chis['ci_upper'] - sacramento_chis['current_prevalence']
    ],
    fmt='none',
    capsize=5
)

plt.xticks(xpos, sacramento_chis['period'], rotation=45)
plt.xlabel('Period (Start–End)')
plt.ylabel('Prevalence (%)')
plt.title('Sacramento County Prevalence by Two-Year Interval')
plt.tight_layout()
plt.show()
```

However, the model still leaves about 61% of the variation unexplained, and the negative slope may point to confounding factors, non‐causal relationships, or limitations in the data.

From these results, we can conclude that there is no strong evidence to support our hypothesis that higher PM2.5 levels are associated with higher lung cancer incidence in the Central Valley. 

### Differences in Air Quality in Californian Regions

To analyze the differences in air quality between the Central Valley and other regions in California, we will compare the PM2.5 levels across different counties. We will focus on the counties in the Central Valley and compare them to counties in other regions of California, such as the coastal and mountainous areas.

```python
all_pm_data
```

We group regions of California into three categories: Valley, Coastal, and Highland.

```python
terrain_classification = {
    'Mountainous': (
        region_classification['shasta_cascade']
        + region_classification['sierra_nevada_&_foothills']
    ),
    'Coastal': (
        region_classification['san_francisco_bay_area']
        + region_classification['central_coast']
        + region_classification['north_coast']
        + region_classification['southern_california']
    ),
    'Valley': region_classification['central_valley'],
}
```

Our null hypothesis states that the average PM2.5 levels are equal between mountainous regions and the Central Valley. 

```python
from scipy.stats import ttest_ind

df = all_pm_data.copy()

# to map each county to its terrain category
lookup = {
    county: terrain
    for terrain, counties in terrain_classification.items()
    for county in counties
}
df['terrain'] = df['location'].map(lookup)

mountain_data = df.loc[df['terrain']=='Mountainous','data'].dropna().astype(float)
valley_data = df.loc[df['terrain']=='Valley','data'].dropna().astype(float)

t_stat, p_val = ttest_ind(mountain_data, valley_data, equal_var=False)

print(f"Mountain vs. Valley PM2.5:\n"
      f"t-statistic = {t_stat:.3f}\n"
      f"p-value = {p_val:.3f}")
```

```python
# Histograms with KDE
plt.figure(figsize=(12, 6))
sns.histplot(mountain_data, label='Mountainous', kde=True, stat='density', alpha=0.5)
sns.histplot(valley_data,    label='Valley',      kde=True, stat='density', alpha=0.5)
plt.xlabel('PM2.5 (µg/m³)')
plt.ylabel('Density')
plt.title('PM2.5 Distribution by Terrain')
plt.legend()
plt.tight_layout()
plt.show()

# Boxplots
plt.figure(figsize=(6, 6))
plot_df = (
    pd.DataFrame({
        'PM2.5': pd.concat([mountain_data, valley_data], ignore_index=True),
        'Terrain': ['Mountainous'] * len(mountain_data) + ['Valley'] * len(valley_data)
    })
)
sns.boxplot(x='Terrain', y='PM2.5', data=plot_df)
plt.xlabel('Terrain')
plt.ylabel('PM2.5 (µg/m³)')
plt.title('Boxplot of PM2.5 by Terrain')
plt.tight_layout()
plt.show()
```

When comparing the average PM2.5 levels in mountainous regions to the Central Valley, we find a negative t-statistic of -4.190, indicating that the average PM2.5 levels in mountainous regions is lower than that of the Central Valley- approximately 4.2 standard deviations lower. 

Additionally, the p-value indicates that there's less than very small chance that we would observe a difference like this if the null hypothesis were true. 

Therefore, we reject the null hypothesis and conclude that the average PM2.5 levels in mountainous regions are significantly lower than that of the Central Valley, p<.001.

Similarly, we can compare the average PM2.5 levels in coastal regions to the Central Valley. 

```python
coastal_data = df.loc[df['terrain']=='Coastal','data'].dropna().astype(float)
valley_data = df.loc[df['terrain']=='Valley','data'].dropna().astype(float)

t_stat, p_val = ttest_ind(coastal_data, valley_data, equal_var=False)

print(f"Coastal vs. Valley PM2.5:\n"
      f"t-statistic = {t_stat:.3f}\n"
      f"p-value = {p_val:.3f}")
```

```python
# Histograms with KDE
plt.figure(figsize=(12, 6))
sns.histplot(coastal_data, label='Coastal', kde=True, stat='density', alpha=0.5)
sns.histplot(valley_data,  label='Valley',  kde=True, stat='density', alpha=0.5)
plt.xlabel('PM2.5 (µg/m³)')
plt.ylabel('Density')
plt.title('PM2.5 Distribution: Coastal vs. Valley')
plt.legend()
plt.tight_layout()
plt.show()

# Boxplots
plt.figure(figsize=(6, 6))
plot_df = pd.DataFrame({
    'PM2.5': pd.concat([coastal_data, valley_data], ignore_index=True),
    'Terrain': ['Coastal'] * len(coastal_data) + ['Valley'] * len(valley_data)
})
sns.boxplot(x='Terrain', y='PM2.5', data=plot_df)
plt.xlabel('Terrain')
plt.ylabel('PM2.5 (µg/m³)')
plt.title('Boxplot of PM2.5: Coastal vs. Valley')
plt.tight_layout()
plt.show()
```

When comparing the average PM2.5 levels in coastal regions to the Central Valley, we find a negative t-statistic of -9.907, indicating that the average PM2.5 levels in coastal regions are lower than that of the Central Valley- approximately 9.9 standard deviations lower. The p-value indicates that there's less than very small chance that we would observe a difference like this if the null hypothesis were true. 

Therefore, we reject the null hypothesis and conclude that the average PM2.5 levels in coastal regions are significantly lower than those of the Central Valley, p<.001.

# Ethics & Privacy

**Data Access and Privacy Constraints** 

A constraint we faced was the lack of access to fine-grain, individual-level health data, which is often protected due to privacy laws like HIPAA. While this is necessary to safeguard personal information, it also limits the depth of analysis we can perform. Because of this, we relied on aggregated county-level averages for both asthma prevalence and pollution exposure. While such data enables general regional insights, it prevents an analysis of how specific demographics, such as children or low-income residents, are uniquely affected. We acknowledged this in our interpretation, avoiding overgeneralization or assumptions about individual outcomes. 

**Data Collection and Representation** 

Air pollution monitoring is typically more concentrated in urban and industrial areas, leaving rural or underfunded counties less represented. This can lead to an underestimate of pollution exposure in areas that lack sensors. Likewise, the asthma data includes notes about statistical instability in smaller counties, which could affect the reliability of those estimates. By noting these limitations in our analysis and visualizations, we aimed to avoid misrepresentation or false precision. 

**Framing and Use of the Data** 

We intentionally framed our research around systemic and environmental factors, rather than placing responsibility on individual behavior. Asthma rates are influenced by proximity to freeways, agricultural zones, and industrial sites, not simply personal health decisions. This framing helps shift the conversation toward environmental safety/justice and public policy, rather than reinforcing stigmatizing narratives. 

**Bias Detection and Communication** 

To address potential biases, we examined the metadata from our datasets for completeness, then flagged counties with unstable estimates, and carefully matched pollution data with health outcomes by region and year. During our analysis, we avoided using the data to make direct predictions about individuals or unsupported claims about causality. In our final presentation, we emphasized transparency about these limits and focused on the broader trends we could confidently support. 

**Equity, Privacy, and Ethical Impact** 

All data used was publicly available and anonymized, minimizing direct privacy risks. However, ethical concerns still exist around interpretation and impact. For example, presenting certain counties as “high risk” without context could unintentionally reinforce negative stereotypes towards that county. We tried to balance this by discussing how social, economic, and environmental policies shape exposure, and by advocating for better air quality monitoring and healthcare access in vulnerable regions.
