# Research Question

Based on some personal experiences, chronic lung diseases such as allergic rhinitis (seasonal allergies), asthma, and lung cancer have been pervasive to citizens' quality of life, especially in the California Central Valley. There seems to exist a correlation between air quality in the Central Valley and respiratory illness. How does air quality correlate with allergic rhinitis, asthma, and lung cancer diagnoses over time in the central valley of California? Furthermore, how does elevation levels and geographic locations of regions in California affect air quality?

## Background and Prior Work

**Background**

The Central Valley of California is a region known not only for its agricultural productivity but also for its persistent air quality issues. Enclosed by mountain ranges and densely populated with farming and industrial activity, the area often experiences high concentrations of air pollutants, such as fine particulate matter (PM2.5). The matter accumulates in the small airways of the lungs, and irritants cause genetic changes responsible for both benign and malignant long-term lung diseases <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1).

Over the course of the decade, more research has been conducted on the potential connection between deteriorating air quality and increased respiratory health problems among residents, including allergy, asthma, and lung cancer diagnoses, especially in Sacramento Valley and San Joaquin Valley <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). From 2015 to 2023, concerns have grown about how this compromised air quality may be affecting public health—particularly the prevalence of seasonal allergies (formally known as allergic rhinitis), asthma, and lung cancer.

**Previous work**

"The Growing Impact of Air Quality on Lung-Related Illness" published in the Journal of Thoracic Disease in 2023, synthesizes findings from studies conducted between 1992 and 2023. The article’s key findings include the effects of particulate matter (PM2.5) on lung function, as epigenetic changes are primarily responsible for bening and malignant lung diseases. The article also found that urban areas experience worse air quality due to higher industrial activity. The article only discusses air quality as a general, but this project aims to look at a connection between elevation and air quality.

Meng, Wilhelm, Lombardi, Balmes, and Ritz (2010)’s "Outdoor Air Pollution and Uncontrolled Asthma in the San Joaquin Valley, California" examines the link between air pollution and asthma in the region. According to the California Health Interview Survey, around 1,502 San Joaquin Valley individuals reported having asthma. They discovered that those living in high-pollution areas were more likely to have asthma symptoms and that poor air quality exacerbated asthma symptoms and increased the number of asthma diagnoses. Overall, it supports our focus on air quality and the effects it has on people's health, assisting us in explaining why many migrate away from these locations since it has only caused them damage, and their desire to get away from it. However, the findings of the study only relate to asthma. The project continues the study by generalizing findings to other respiratory ailments too, such as allergic rhintis and lung cancer.

### California Health Maps - Cancer Registry Data
The California Health Maps Cancer Registry dataset provides age-adjusted incident rates for different cancer types, broken down by county and demographic variables. Age-adjusted incident rates is a standardized measure that accounts age differences in population, and make it feasable to compare cancer rates between regions with different age distributions (urban versus rural areas). Data cleaning for this dataset involves filtering for lung cancer cases and aggregating by California county for geographic analysis. The most important variables for our analysis are Counties, Cities, Cancer, and AAIR.

### CHIS Data - Current Asthma Prevalence by County 2015 - 2022
This dataset, derived from the California Health Interview Survey (CHIS), provides estimates of the percentage of Californians who have been diagnosed with asthma or experienced an asthma episode in the last twelve months. The data is offered by county and statewide levels. Key features that we will be taking into account in our data analysis is the timeframe, reported by two year intervals, all California counties that encompass the Central Valley, and the percentage of Californians with asthma symptoms.

This dataset contains PM2.5 air quality data measured across different counties in California, specifically in Central Valley. ### USGS Lidar Point Cloud Data – AWS Open Registry
This dataset contains high-resolution Light Detection and Ranging (LiDAR) point cloud data collected by the U.S. Geological Survey (USGS). ## CHIS Data - Current Asthma Prevalence by County 2015 - 2022

```python
import pandas as pd
import matplotlib.pyplot as plt

chis_data = pd.read_csv('./data/current-asthma-prevalence-by-county-2015_2022.csv')
chis_data.columns = chis_data.columns.str.lower() # Normalize column names to lowercase
chis_data.columns = chis_data.columns.str.replace(' ', '_')
```

```python
# California isn't a county, so we remove it.
chis_data = chis_data[chis_data['county'] != 'California']
chis_data["county"].unique()
```

```python
chis_data = chis_data[chis_data['age_group'] == 'All ages']

chis_data
```

We have the standardize the data in each column

```python
chis_data["county"] = chis_data["county"].str.replace(" County", "", regex=False)
chis_data["county"] = chis_data["county"].str.lower()
chis_data["county"] = chis_data["county"].str.replace(" ", "_", regex=False)

chis_data["county"].unique() 
```

```python
# split the "years" column into start_year and end_year
chis_data[['start_year', 'end_year']] = chis_data['years'].str.split('-', expand=True)

chis_data["start_year"] = pd.to_numeric(chis_data["start_year"], errors='coerce')
chis_data["end_year"] = pd.to_numeric(chis_data["end_year"], errors='coerce')

# remove the "years" column
chis_data = chis_data.drop(columns=['years'])

display(chis_data["start_year"].unique())
display(chis_data["end_year"].unique())
```

```python
chis_data["strata"] = chis_data["strata"].str.lower()
chis_data["strata"] = chis_data["strata"].str.replace(" ", "_", regex=False)

chis_data["strata"].unique() # we could drop this
```

```python
chis_data["age_group"] = chis_data["age_group"].str.lower()
chis_data["age_group"] = chis_data["age_group"].str.replace(" ", "_", regex=False)

chis_data["age_group"].unique() # we could drop this
```

```python
# chis_data["95%_confidence_interval"]

# make a new column for lower and upper bounds of the confidence interval
chis_data['95%_confidence_interval'] = chis_data['95%_confidence_interval'].str.replace('(', '', regex=False)
chis_data['95%_confidence_interval'] = chis_data['95%_confidence_interval'].str.replace(')', '', regex=False)
chis_data[['ci_lower', 'ci_upper']] = chis_data['95%_confidence_interval'].str.split('-', expand=True)
chis_data["ci_lower"] = pd.to_numeric(chis_data["ci_lower"], errors='coerce')
chis_data["ci_upper"] = pd.to_numeric(chis_data["ci_upper"], errors='coerce')
# remove the "95%_confidence_interval" column
chis_data = chis_data.drop(columns=['95%_confidence_interval'])

# remove nan values in the ci_lower and ci_upper columns
chis_data = chis_data.dropna(subset=['ci_lower', 'ci_upper'])
```

```python
print(f"\nNumber of observations: {chis_data.shape[0]}")
print(f"Number of variables: {chis_data.shape[1]}")
```

### Analyzing Asthma Prevalence across counties in the Central Valley over time

Now, let's look at the asthma data and compare the prevalence of asthma across different counties in the Central Valley. First, we filter chis_data and pm_data for the counties in the Central Valley they share.

```python
chis_counties = chis_data["county"].unique()
pm_counties = pm_data["location"].unique()

# find the intersection of the two lists
common_counties = set(pm_counties).intersection(set(chis_counties))
print("Common counties between PM data and CHIS data:")
print(common_counties)
```

In pm_data, we are given the "timeframe" of when the PM2.5 levels were measured, and in chis_data, we are given a two-year interval of when the asthma data was collected. For the sake of simplicity, we will use the first year of the two-year interval in chis_data to match with the timeframe in pm_data.

```python
chis_years = chis_data["start_year"].unique().tolist()
pm_years = pm_data["timeframe"].unique().tolist()

# find the intersection of the two lists
common_years = set(pm_years).intersection(set(chis_years))
print("Common years between PM data and CHIS data:")
print(common_years)
```

```python
common_chis_data = chis_data[chis_data["county"].isin(common_counties) & chis_data["start_year"].isin(common_years)]
common_pm_data = pm_data[pm_data["location"].isin(common_counties) & pm_data["timeframe"].isin(common_years)]
```

```python
# lets looks at sacramento
sacramento_chis = common_chis_data[common_chis_data["county"] == "sacramento"]
sacramento_pm = common_pm_data[common_pm_data["location"] == "sacramento"]

display(sacramento_chis)
display(sacramento_pm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sacramento_chis['period'] = (sacramento_chis['start_year'].astype(str)+ '-' + sacramento_chis['end_year'].astype(str))

plt.figure(figsize=(8, 5))
sns.lineplot(
    data=sacramento_chis,
    x='period',
    y='current_prevalence',
    marker='o'
)

xpos = range(len(sacramento_chis))
plt.errorbar(
    xpos,
    sacramento_chis['current_prevalence'],
    yerr=[
        sacramento_chis['current_prevalence'] - sacramento_chis['ci_lower'],
        sacramento_chis['ci_upper'] - sacramento_chis['current_prevalence']
    ],
    fmt='none',
    capsize=5
)

plt.xticks(xpos, sacramento_chis['period'], rotation=45)
plt.xlabel('Period (Start–End)')
plt.ylabel('Prevalence (%)')
plt.title('Sacramento County Prevalence by Two-Year Interval')
plt.tight_layout()
plt.show()
```

```python
# plot the sacramento PM data
plt.figure(figsize=(8, 5))
sns.lineplot(
    data=sacramento_pm,
    x='timeframe',
    y='data',
    marker='o'
)
plt.xlabel('Year')
plt.ylabel('Average Particulate Matter Concentration')
plt.title('Sacramento County Average Particulate Matter Concentration Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

In order to make pm_data and chis_data comparable, we want to bin the PM2.5 levels into two-year intervals. This will allow us to see the average PM2.5 levels in each county over the years. Let's look at all the counties in the Central Valley and their average PM2.5 levels over the years.

```python
central_valley_chis_data = common_chis_data
central_valley_pm_data = common_pm_data

central_valley_chis_data = central_valley_chis_data.drop(columns=['counties_grouped', 'comment', 'strata', 'age_group', 'ci_lower', 'ci_upper', 'end_year'])

central_valley_pm_data = central_valley_pm_data.drop(columns=['dataformat', 'locationtype', 'fipscode'])

display(central_valley_pm_data)
display(central_valley_chis_data)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=central_valley_pm_data,
    x='timeframe',
    y='data',
    hue='location',
    marker='o'
)
plt.xlabel('Year')
plt.ylabel('Average Particulate Matter Concentration (µg/m³)')
plt.title('Average Particulate Matter Concentration (µg/m³) Over Time by County in Central Valley')
plt.legend(title='County', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=central_valley_chis_data,
    x='start_year',
    y='current_prevalence',
    hue='county',
    marker='o'
)
plt.xlabel('Year')
plt.ylabel('Current Prevalence (%)')
plt.title('Current Prevalence of Asthma Over Time by County in Central Valley')
plt.legend(title='County', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# create a df with the average particulate matter concentration and current prevalence of asthma by county
central_valley_combined = pd.merge(
    central_valley_chis_data,
    central_valley_pm_data,
    left_on=['county', 'start_year'],
    right_on=['location', 'timeframe'],
    how='inner'
)
central_valley_combined.drop(columns=['location', 'start_year'], inplace=True)
central_valley_combined
```

```python
# !pip install statsmodels

import statsmodels
import statsmodels.formula.api as smf

model = smf.ols(formula='current_prevalence ~ C(data)',data=central_valley_combined).fit()

res_1 = model.summary()
display(res_1)
```

When looking at the relationship between the prevalence of asthma and the PM2.5 of the county, we see a moderate positive correlation between the two variables. The regression analysis appears to indicate that a higher concentration of PM2.5 could predict a higher prevalence in asthma.

Our biggest limitation when conducting the regression analysis was the timeline of our CHIS data. The CHIS data only had data from 2015 to 2021, which wasn't enough time to see the greater picture of PM2.5 concentration over a longer amount of time, but it had sufficient enough data to see a moderate pattern in light of recent years. Regardless, it was still insightful to see that there is, in fact, a correlation between PM2.5 concentration and prevalence of asthma in the last ten years.
